```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import MinMaxScaler

warnings.filterwarnings('ignore')
# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
from scipy.stats import chi2_contingency, ttest_ind
```

```python
df = pd.read_excel('人员培训+被试采集.xlsx', sheet_name=1)
df = df.loc[:32, :]
df['性别'] = df['性别'].str.strip()  # 去掉首尾空格
df
```

```python
# 假设你的数据叫 df
# 1. 描述性统计
print("性别人数：")
print(pd.crosstab(df['组别'], df['性别']))  # 各组性别人数

print("\n年龄均值和标准差：")
print(df.groupby('组别')['年龄'].agg(['mean', 'std', 'count']))

# 2. 性别组间差异 → 卡方检验
cont_table = pd.crosstab(df['组别'], df['性别'])
chi2, p_chi, dof, expected = chi2_contingency(cont_table)
print(f"\n性别分布卡方检验：chi2={chi2:.2f}, p={p_chi:.3f}")

# 3. 年龄组间差异 → t检验
age_alcohol = df.loc[df['组别'] == 'A', '年龄']
age_control = df.loc[df['组别'] == 'B', '年龄']
t, p_t = ttest_ind(age_alcohol, age_control, equal_var=False)  # Welch检验
print(f"年龄t检验：t={t:.2f}, p={p_t:.3f}")
```

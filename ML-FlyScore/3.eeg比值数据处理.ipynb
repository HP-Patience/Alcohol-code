```python
# 你的文件路径字典
file_paths = {
    "A_alpVShi_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\A_alpVShi_ratio.mat",
    "A_alpVSlow_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\A_alpVSlow_ratio.mat",
    "A_theVShi_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\A_theVShi_ratio.mat",

    "A_theVSlow_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\A_theVSlow_ratio.mat",
    "B_alpVShi_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\B_alpVShi_ratio.mat",
    "B_alpVSlow_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\B_alpVSlow_ratio.mat",
    "B_theVShi_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\B_theVShi_ratio.mat",
    "B_theVSlow_ratio": r"E:\pycharm all files\眼动数据处理\相关性分析\数据\原始数据\eeg比值&熵\比值\B_theVSlow_ratio.mat",
}

# 3 channels 的标签
channels = ["Fz", "Cz", "Pz"]

# 最终保存所有行
rows = []

for key, path in file_paths.items():
    with h5py.File(path, "r") as f:
        data = f[key][()]  # shape = (3,16,7)

    # key 信息解析，例如：A_alpVShi_ratio
    group = key[0]  # A / B
    ratio_type = key[2:-6]  # 去掉 "A_" 和 "_ratio"

    # 展开数据
    for ch_idx in range(3):
        for subj in range(16):
            for sess in range(7):
                value = data[ch_idx, subj, sess]

                # ⭐ 合并新指标 metric
                metric = f"{channels[ch_idx]}_{ratio_type}"
                rows.append({
                    "group": group,
                    "metric": metric,
                    "subject": subj + 1,
                    "session": sess + 1,
                    "value": float(value)
                })

# 转为 DataFrame
df = pd.DataFrame(rows)

df
```

```python
# A组与B组姓名
names_A = ["曾奇", "王榕榕", "徐梓军", "徐文婷", "达格吉", "李天琦", "王威", "张渊博",
           "黄博文", "朱祖恩", "刘子皓", "李嘉文", "徐子焰", "李国荣", "杨珊珊", "付瑞晗"]

names_B = ["杜佳鑫", "冯科嘉", "陈妍", "胡浩男", "王子铭", "程腾宇", "朱娇娇", "祖丽米热",
           "毛华灏", "郭浚杰", "周鑫颜", "肖雨晨", "冯晓娅", "刘勇杰", "祝鹏程", "董嘉乐"]

# 构建映射字典
map_A = {i + 1: names_A[i] for i in range(len(names_A))}  # 1–16
map_B = {i + 1: names_B[i] for i in range(len(names_B))}  # 1–16

# 根据 group 替换 subject → name
df["name"] = df.apply(lambda row:
                      map_A[row["subject"]] if row["group"] == "A"
                      else map_B[row["subject"]],
                      axis=1)

# 如不需要原 subject 列，可删除
df = df.drop(columns=["subject"])
```

```python
# pivot wider: metric 作为新列
df_wide = df.pivot_table(
    index=["name", "group", "session"],
    columns="metric",
    values="value"
).reset_index()

# 可选：列名扁平化（防止 MultiIndex）
df_wide.columns.name = None

df_wide
```

```python
df_wide.group = df_wide.group.replace({"A": "Alcohol", "B": "Control"})
df_wide.columns = ['受试者', '组别', 'session', 'Cz_alpVShi', 'Cz_alpVSlow', 'Cz_theVShi',
                   'Cz_theVSlow', 'Fz_alpVShi', 'Fz_alpVSlow', 'Fz_theVShi', 'Fz_theVSlow',
                   'Pz_alpVShi', 'Pz_alpVSlow', 'Pz_theVShi', 'Pz_theVSlow']
df_wide
```

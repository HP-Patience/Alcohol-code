```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.preprocessing import PowerTransformer

warnings.filterwarnings('ignore')
# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 读取数据
df = pd.read_excel(
    r'E:\pycharm all files\眼动数据处理\完整眼动数据分析\眼动数据插值处理文件.xlsx')
df
```

```python
len(df.被试者.unique())
```

```python
# 去掉最后一列
df = df.iloc[:, :-1]
# 重置列名
df.columns = ['阶段', 'AOI转换次数', '静态注释熵(SGE)', '眼跳注视熵(GTE)', '组别', '性别', '受试者', '飞行天数']
# 重命名阶段
df['阶段'] = df['阶段'].replace(['第1次转弯', '第2次转弯', '第3次转弯', '第4次转弯'], '转弯')
df['阶段'] = df['阶段'].replace('起飞阶段', '起飞')
df['阶段'] = df['阶段'].replace('降落阶段', '降落')
df['阶段'] = df['阶段'].replace('巡航阶段', '巡航')
# 替换组别名称
df["组别"] = df["组别"].replace({
    "A": "Alcohol",
    "B": "Control"
})

df
```

```python
# 查看缺失值
print('缺失值个数：', df.isnull().sum().sum())
# 查看异常值
df1 = df[df['AOI转换次数'].isnull()]
df1
```

```python
# 删除缺失值
df = df.dropna()
print(df.isnull().sum())
```

```python
df.to_excel(r'眼动数据格式对齐后文件.xlsx', index=False)
```

```python
# 设定绘图风格
sns.set(style="whitegrid", font="SimHei")  # 支持中文显示

# 三个阶段
phases = df["阶段"].unique()
label = 'AOI转换次数'
# 建立图表
for phase in phases:
    plt.figure(figsize=(6, 4))

    # if phase != "降落":
    #     continue
    # 筛选该阶段数据
    sub_df = df[df["阶段"] == phase]

    sns.pointplot(
        data=sub_df,
        x="飞行天数",
        y=label,
        hue="组别",
        # errorbar=("ci", 95), # 误差线显示 95% 置信区间 # 或 ("se", 1)
        errorbar=("se", 1),  # 画标准误 误差线显示 ±1 标准误
        dodge=True,
        join=True,
        capsize=0.2,  # 控制小横线长度 (相对比例)
        errwidth=1.2,  # 误差线粗细，可调
        markersize=4
    )

    plt.title(f"{phase}阶段：{label}随飞行天数变化")
    plt.xlabel("飞行天数")
    plt.ylabel(label)
    plt.legend(title="组别")
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pathlib import Path

warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 配置根目录
ROOT_DIR = Path(r"E:\pycharm all files\眼动数据处理\Data")

# 阶段文件映射
PHASE_FILES = {
    "1.qf.xlsx": "起飞阶段",
    "2.zw1.xlsx": "第1次转弯",
    "3.zw2.xlsx": "第2次转弯",
    "4.xh.xlsx": "巡航阶段",
    "5.zw3.xlsx": "第3次转弯",
    "6.zw4.xlsx": "第4次转弯",
    "7.jl.xlsx": "降落阶段"
}

# 无效数据记录
INVALID_DATA = {
    ('A', '女', '杨珊珊'): [3],
    ('A', '男', '徐梓军'): [1, 2, 6],
    ('A', '男', '曾奇'): [1],
    ('A', '男', '朱祖恩'): [2, 4],
    ('A', '女', '王蓉蓉'): [7],
    ('B', '男', '冯科嘉'): [1],
    ('B', '男', '毛华浩'): [6],
    ('B', '男', '王子铭'): [6],
    ('B', '男', '祝鹏程'): [3],
    ('B', '男', '程腾宇'): [1, 2]
}
```

```python
def calculate_eye_duration(root_dir):
    total_duration_sec = 0
    durations = []

    # 遍历组别
    for group_dir in root_dir.iterdir():
        if not group_dir.is_dir() or group_dir.name not in ['A', 'B']:
            continue

        for gender_dir in group_dir.iterdir():
            if not gender_dir.is_dir() or gender_dir.name not in ['男', '女']:
                continue

            for participant_dir in gender_dir.iterdir():
                if not participant_dir.is_dir():
                    continue

                for day in range(1, 8):
                    day_dir = participant_dir / str(day)

                    # 跳过无效数据
                    if (group_dir.name, gender_dir.name, participant_dir.name) in INVALID_DATA:
                        if day in INVALID_DATA[(group_dir.name, gender_dir.name, participant_dir.name)]:
                            continue

                    if not day_dir.exists() or not day_dir.is_dir():
                        continue

                    print(f"处理第{day}天数据")

                    for phase_file, phase_name in PHASE_FILES.items():
                        file_path = day_dir / phase_file
                        if not file_path.exists():
                            continue

                        try:
                            df = pd.read_excel(file_path)
                            if df.shape[1] < 4:
                                continue

                            # 第4列当作秒数
                            time_series = pd.to_numeric(df.iloc[:, 3], errors='coerce').dropna()

                            if len(time_series) > 1:
                                duration_sec = time_series.max() - time_series.min()
                            else:
                                duration_sec = 0

                            total_duration_sec += duration_sec

                            durations.append({
                                "组别": group_dir.name,
                                "性别": gender_dir.name,
                                "被试者": participant_dir.name,
                                "天数": day,
                                "阶段": phase_name,
                                "文件路径": str(file_path),
                                "时长(秒)": duration_sec,
                                "时长(毫秒)": duration_sec * 1000
                            })

                        except Exception as e:
                            print(f"处理文件 {file_path} 时出错: {e}")

    durations_df = pd.DataFrame(durations)

    # 打印总时长
    print(f"\n所有被试所有阶段的总眼动数据时长: {total_duration_sec:.2f} 秒")

    # # 打印每文件明细
    # print("\n==== 每个文件的时长明细 ====")
    # print(durations_df)

    # 每阶段均值±标准差
    if not durations_df.empty:
        summary = []
        for phase, group in durations_df.groupby("阶段"):
            mean_sec = group["时长(秒)"].mean()
            std_sec = group["时长(秒)"].std()
            mean_ms = group["时长(毫秒)"].mean()
            std_ms = group["时长(毫秒)"].std()

            summary.append({
                "阶段": phase,
                "mean±std(秒)": f"{mean_sec:.2f} ± {std_sec:.2f}",
                "mean±std(毫秒)": f"{mean_ms:.0f} ± {std_ms:.0f}"
            })

        summary_df = pd.DataFrame(summary)
        print("\n==== 每个阶段的平均值±标准差 ====")
        print(summary_df)

    return durations_df, summary_df
```

```python
if __name__ == "__main__":
    durations_df, summary_df = calculate_eye_duration(ROOT_DIR)
```
